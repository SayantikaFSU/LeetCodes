'Binary Tree Node' is a function which is applied to create the following structure

Each node in a Binary Tree has three parts:

* Data

```python
class BinaryTreeNode:
    def __init__(self, data=None):
        self.data = data
        self.left = None
        self.right = None
```

```python
# form the tree:

## Tree :
root=BinaryTreeNode(1)
node1=BinaryTreeNode(2)
node2=BinaryTreeNode(3)
node3=BinaryTreeNode(4)
node4=BinaryTreeNode(5)
node6= BinaryTreeNode(6)

root.left=node1
root.right=node2
node1.left=node3
node1.right=node4
node2.right=node6


# Preorder Traversal :

def preOrder_traversal(node):
    if node is None:
        return node
    print(node.data,end=" ")
    preOrder_traversal(node.left)
    preOrder_traversal(node.right)

preOrder_traversal(root)

print('\n \n')
# Post Order Traversal:
def postOrder_traversal(node):
    if node is None:
        return node
    postOrder_traversal(node.left)
    postOrder_traversal(node.right)
    print(node.data,end=" ")

postOrder_traversal(root)

```

Problem 3: Insertion in Binary Tree
*************************************************
Inserting elements means add a new node into the binary tree. As we know that there is no such ordering of elements in the binary tree, So we do not have to worry about the ordering of node in the binary tree. We would first creates a root node in case of empty tree. Then subsequent insertions involve iteratively searching for an empty place at each level of the tree. When an empty left or right child is found then new node is inserted there. By convention, insertion always starts with the left child node.




image.png

```python
# form the tree:

## Tree :
root=BinaryTreeNode(2)
node1=BinaryTreeNode(3)
node2=BinaryTreeNode(4)
node3=BinaryTreeNode(5)


root.left=node1
root.right=node2
node1.left=node3

print('Pre Order Traversal')
preOrder_traversal(root)

print('\n Post Order Traversal')
postOrder_traversal(root)
```

```python
from collections import deque

def insert_at_level_order(root,new_data):

    if root is None: # if the tree at all doesnot exists
        root=BinaryTreeNode(new_data)

    q=deque([root]) # create a deque at the root
    while q:
            # create a temp left position
            temp=q.popleft()
            if temp.left is None: # if that place is empty insert the new_data and break
                temp.left=BinaryTreeNode(new_data)
                break
            else:
                print(temp.left.data)
                q.append(temp.left) # to iterate over q

            if temp.right is None: # if that place is empty insert the new_data and break
                temp.right=BinaryTreeNode(new_data)
                break

            else:
                print(temp.right.data)
                q.append(temp.right) # to iterate over q

    return  root

```

Problem 4: Searching in Binary Tree
************************************************************
Searching for a value in a binary tree means looking through the tree to find a node that has that value. Since binary trees do not have a specific order like binary search trees, we typically use any traversal method to search. The most common methods are depth-first search (DFS) and breadth-first search (BFS). In DFS, we start from the root and explore the depth nodes first. In BFS, we explore all the nodes at the present depth level before moving on to the nodes at the next level. We continue this process until we either find the node with the desired value or reach the end of the tree. If the tree is empty or the value isnt found after exploring all possibilities, we conclude that the value does not exist in the tree.

Here is the implementation of searching in a binary tree using Depth-First Search (DFS)

```python
# Function to search for a value in the binary tree using DFS
def searchDFS(root, value):
    # Base case: If the tree is empty or we've reached a leaf node
    if root is None:
        return False
    # If the node's data is equal to the value we are searching for
    if root.data == value:
        return True
    # Recursively search in the left and right subtrees
    left_res = searchDFS(root.left, value)
    right_res = searchDFS(root.right, value)

    return left_res or right_res

if __name__ == "__main__":
    root = BinaryTreeNode(2)
    root.left = BinaryTreeNode(3)
    root.right = BinaryTreeNode(4)
    root.left.left = BinaryTreeNode(5)
    root.left.right = BinaryTreeNode(6)

    value = 6
    if searchDFS(root, value):
        print(f"{value} is found in the binary tree")
    else:
        print(f"{value} is not found in the binary tree")
```

Problem 5: Test if a Tree is Height Balanced:
***************************************************
* A tree is height balanced if for each node , the difference in height of left and right subtrees is atmost 1.

* height of a node n = number of nodes in the search path of n-> root, not including n



```python
# Find height of a tree : node -> [node.left    node.right]
def height(node):
    if node is None:
        return 0
    else:
        h = 1 + max(height(node.left), height(node.right)) # recursive
        return h

def is_balanced(root):
    if root is None:
        print('Tree is balanced')
    else:
       r_h = height(root.right)  # height(right_subtree)
       l_h = height(root.left)
       if abs(r_h - l_h) <= 1:
           print('Tree is balanced')
       else:
           print('Tree is not balanced')
    return r_h, l_h

if __name__ == "__main__":

       # tree: 2-> [3->[5->6]  4]

        # Step 1: Allocate the nodes
        root=BinaryTreeNode(2)
        child1=BinaryTreeNode(3)
        child2=BinaryTreeNode(4)
        child3=BinaryTreeNode(5)
        child4=BinaryTreeNode(6)

        # Step 2: Make Connections
        root.left=child1
        root.right=child2
        child1.left=child3
        child3.right=child4

        # Check
        r_h, l_h=is_balanced(root)
        print('r_h, l_h = %d, %d'%(r_h,l_h))
```

Problem 6: If a binary tree is symmetric
********************************************************
Symmetric : if a vertical line is drawn through root then, left subtree = right subtree

```python
def symmetric(tree1, tree2):
    # both empty then symmteric
    if tree1 is None and tree2 is None:
         value=1

    #For two trees to be mirror images, the following three conditions must be true
    #1: Their root node's data must be same
    #2: left subtree of left tree and right subtree of the right tree have to be mirror images
    #3: right subtree of left tree and left subtree of right tree have to be mirror images
    else:
         if tree1.data == tree2.data:
            value= symmetric(tree1.left, tree2.right) and symmetric(tree1.right, tree2.left)
         else:
            value=0

    return value


def is_symmetric(root):
    v= symmetric(root.left, root.right)
    if v==1:
        print('Tree is symmetric')
    else:
        print('Tree is not symmetric')


if __name__ == "__main__":
    # 2-> [3->[5 6]         3->[6 5]]
    root = BinaryTreeNode(2)
    root.left = BinaryTreeNode(3)
    root.right = BinaryTreeNode(3)
    root.left.left = BinaryTreeNode(5)
    root.left.right = BinaryTreeNode(6)
    root.right.left = BinaryTreeNode(6)
    root.right.right = BinaryTreeNode(5)

is_symmetric(root)
```
